## CLUSTRING

```python
def time_change(dataset):
    df = dataset.copy()
    date_columns = df.select_dtypes(include=['datetime64']).columns.tolist()
    df[date_columns] = df[date_columns].astype(str)
    for column in df:
        if df[column].dtype==object:
            df[column]=pd.to_datetime(df[column])
            df[f'{column}_YEAR']=pd.DatetimeIndex(df[column]).year
            df[f'{column}_MONTH']=pd.DatetimeIndex(df[column]).month
            df[f'{column}_DATE']=pd.DatetimeIndex(df[column]).day
            df=df.drop(column,axis=1)
    return df
```

```python
prepared_data=time_change(df)
```

```python
prepared_data.head(2)
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(prepared_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(prepared_data)
data_scaled = pd.DataFrame(data_scaled, columns=prepared_data.columns)
data_scaled.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10, 7)) 
plt.scatter(data_scaled['REVENUE'], data_scaled['TOTAL_ORDERS'], c=cluster.labels_)
```

```python
import scipy.cluster.hierarchy as hc
plt.figure(figsize=(10, 7)) 
plt.title("Dendrograms") 
dend = hc.dendrogram(hc.linkage(data_scaled, method='ward'))
plt.axhline(y=6, color='r', linestyle='--')
```

## RFM

```python
df.head(2)
```

```python
recency_df = df.groupby(by='CustomerID', as_index=False)['FIRST_ORDER_DATE'].max()
recency_df.columns = ['CustomerID','LastPurshaceDate']
recent_date=recency_df.LastPurshaceDate.max()
recency_df['Recency'] = recency_df['LastPurshaceDate'].apply(lambda x: (recent_date- x).days)
recency_df.head()
```

```python
frequency_df = df.drop_duplicates().groupby(by=['CustomerID'], as_index=False)['LATEST_ORDER_DATE'].count()
frequency_df.columns = ['CustomerID', 'Frequency']
frequency_df.head()
```

```python
df['TotalCost'] = df['TOTAL_ORDERS'] * df['AVERAGE_ORDER_VALUE']
monetary_df = df.groupby(by='CustomerID',as_index=False).agg({'TotalCost': 'sum'})
monetary_df.columns = ['CustomerID','Monetary']
monetary_df.head()
```

```python
temp_df = recency_df.merge(frequency_df,on='CustomerID')
temp_df.head()
```

```python
rfm_df = temp_df.merge(monetary_df,on='CustomerID')
# rfm_df.set_index('CustomerID',inplace=True)
rfm_df.head()
```

```python
plt.subplot(3, 1, 1); sns.distplot(rfm_df['Recency'])
plt.subplot(3, 1, 2); sns.distplot(rfm_df['Frequency'])
plt.subplot(3, 1, 3); sns.distplot(rfm_df['Monetary'])
plt.show()
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)

rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100
 
rfm_df.head()
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28 * \
    rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['CustomerID', 'RFM_Score']].head(7)
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] > 4.5, "Top Customers",
    (np.where(rfm_df['RFM_Score'] > 4,"High value Customer",(np.where(rfm_df['RFM_Score'] > 3,"Medium Value Customer",
    np.where(rfm_df['RFM_Score'] > 1.6,'Low Value Customers', 'Lost Customers'))))))
rfm_df[['CustomerID', 'RFM_Score', 'Customer_segment']].head(20)
```
