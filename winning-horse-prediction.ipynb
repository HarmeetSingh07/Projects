```python
# feature engineering
# trainer
trainer_df = data.trainer.value_counts().to_frame('trainer')

d = {range(1,11):1,range(11,51):2,range(50,101):3,range(101,201):4,range(201,301):5,range(301,501):6,range(501,701):7,
                         range(701,1001):8,range(1001,2001):9,range(2001,3369):10}

trainer_df['trainer']=trainer_df['trainer'].apply(lambda x: next((v for k, v in d.items() if x in k), 0))
trainer_df = trainer_df.reset_index()
trainer_df.columns = ['trainer','trainer_count']
data = pd.merge(data,trainer_df,on=['trainer'],how='left')



# driver
driver_df = data.driver.value_counts().to_frame('driver_count').reset_index()

d = {range(1,11):1,range(11,51):2,range(50,101):3,range(101,201):4,range(201,301):5,range(301,501):6,range(501,701):7,
                         range(701,1001):8,range(1001,1501):9,range(1501,1721):10}

driver_df.columns = ['driver','driver_count']
driver_df['driver_count']=driver_df['driver_count'].apply(lambda x: next((v for k, v in d.items() if x in k), 0))
data = pd.merge(data,driver_df,on=['driver'],how='left')
# data.drop(['trainer','driver'],axis=1,inplace=True)


# venue
venue_race_count_df = data.groupby(['venue','horsename'])['horsename'].count().sort_values().to_frame('race_count')
data = pd.merge(data,venue_race_count_df,on=['venue','horsename'],how='left')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer


def num_pipeline_transformer(data):
    numerics = ['float64', 'int64']

    num_attrs = data.select_dtypes(include=numerics)

    num_pipeline = Pipeline([('scaler', RobustScaler())])
    return num_attrs, num_pipeline


def pipeline_transformer(data):
    cat_attrs = list(data.select_dtypes(include=object).columns)
    num_attrs, num_pipeline = num_pipeline_transformer(data)
    full_pipeline = ColumnTransformer([
        ("num", num_pipeline, list(num_attrs)),
        ("cat", OrdinalEncoder(), cat_attrs),
        ])
    prepared_data = full_pipeline.fit_transform(data)
    prepared_data = pd.DataFrame(prepared_data,columns=data.columns) 
    return prepared_data
```

```python
from imblearn.under_sampling import RandomUnderSampler

X = data.drop('place', axis = 1)
y = data.place
prepared_data = pipeline_transformer(X)
print('Before\n')
print(y.value_counts())

rus = RandomUnderSampler(random_state=42,sampling_strategy='majority') 
prepared_data, y = rus.fit_resample(prepared_data, y)
print('After\n')
print(y.value_counts())
```

```python
lst = ['racenumber', 'racename', 'racetitle', 'racedistance', 'horsename',
       'prizemoney', 'trainer', 'driver', 'margin', 'startingodds', 'venue',
       'daycalender', 'handicap', 'placer', 'horseid', 'age', 'colour', 'sire',
       'sex', 'winner', 'firststarter', 'lifetimeruns', 'recentruns',
       'lifetimewins', 'lifetimeseconds', 'lifetimethirds', 'lifetimeplacings',
       'lifetimewinpercent', 'lifetimeplacepercent', 'recentplacepercent',
       'recentwinpercent', 'recentplacings', 'recentthirds',
       'leaderstrikerate', 'start2', 'start3', 'start4', 'marginwinner',
       'samedistance', 'leading', 'deathseat', 'downindistance',
       'trainerlaststart', 'driverlaststart', 'bestwinningmile', 'timelead',
       'standlaststart', 'racetime', 'morningrace', 'eveningrace',
       'timelastrace', 'eveningracelaststart', 'prizemoneylastrace',
       'startingprice1', 'avestake', 'sametrack', 'prizemoneysame',
       'prizemoneyup', 'prizemoneydown', 'tracklaststart',
       'tracknamelaststart', 'trials', 'aveplace', 'countld1', 'countld2',
       'countld3', 'countld4', 'countfront1', 'countfront2', 'countfront3',
       'countfront4', 'countofdeath', 'frontrow', 'secondrow', 'numbertrainer',
       'trainer_count', 'driver_count', 'race_count']
```

```python
prepared_data = prepared_data[lst]
```

```python
def OutlireHandling(imputed_df):
    columns = lst
    for column in columns:
        high = imputed_df[column].mean() + 2.5*imputed_df[column].std()
        low = imputed_df[column].mean() - 2.5*imputed_df[column].std()
        imputed_df.loc[imputed_df[column] > high, column] = high
        imputed_df.loc[imputed_df[column] < low, column] = low
    ModelReady_df = imputed_df
    return ModelReady_df

prepared_data = OutlireHandling(prepared_data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(prepared_data,y,test_size=0.2,random_state=211)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(min_samples_split= 10, criterion= 'entropy', max_features = 48
                                    ,bootstrap= False, max_depth= 15, n_estimators= 100)
forest_clf.fit(X_train, y_train)
forest_clf_cv_scores = cross_val_score(forest_clf,X_train, y_train, scoring='accuracy', cv = 10)

forest_clf_accuracy_scores = forest_clf_cv_scores
forest_clf_accuracy_scores.mean()
```

```python
from sklearn.metrics import accuracy_score

place_predictions = forest_clf.predict(X_test)
test_score = accuracy_score(y_test,place_predictions)
test_score
```
